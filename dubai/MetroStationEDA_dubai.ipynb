```python
import pandas as pd

# Load the dataset
file_path = 'Finall_dubai.csv' 
metro_dataset = pd.read_csv(file_path, encoding='latin1')

# Display the first few rows to inspect the data
print(metro_dataset.head())
```

```python
import pandas as pd
from IPython.display import display

# Getting the general information as a DataFrame
info_df = pd.DataFrame(metro_dataset.dtypes).rename(columns={0: 'Dtype'})
info_df['Non-Null Count'] = metro_dataset.count()
info_df = info_df.reset_index().rename(columns={'index': 'Column'})
print("General Information:")
display(info_df)

# Summary statistics for numerical columns
print("\nSummary Statistics for Numerical Columns:")
display(metro_dataset.describe())

# Check for missing values
missing_values_df = metro_dataset.isnull().sum().reset_index()
missing_values_df.columns = ['Column', 'Missing Values']
print("\nMissing Values in Each Column:")
display(missing_values_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.imshow(metro_dataset.isna(), aspect="auto", interpolation="nearest", cmap="gray")

# Set the x-tick labels to the column names
plt.xticks(ticks=range(len(metro_dataset.columns)), labels=metro_dataset.columns, rotation=90)

# Labeling the axes
plt.xlabel("Columns")
plt.ylabel("Sample Number")

# Display the plot
plt.show()
```

**Summary**

This chart represents a missing data matrix, where rows correspond to samples and columns to dataset features. The black areas indicate the presence of data, while white or blank spaces (if any) would denote missing values. From the chart, all columns appear fully populated with no visible missing data. so that mean we dont have any missing vlaue.

```python
import matplotlib.pyplot as plt

# Define colors for each attraction type
attraction_colors = {
    "Entertainment": "blue",
    "Workplace": "green",
    "Health": "red",
    "Education": "purple",
    "Other": "orange"
}

# Define the attraction columns for each type
attraction_columns = {
    "Entertainment": ["Entertainment_Attraction", "Entertainment_Latitude", "Entertainment_Longitude"],
    "Workplace": ["Workplace_Attraction", "Workplace_Latitude", "Workplace_Longitude"],
    "Health": ["Health_Attraction", "Health_Latitude", "Health_Longitude"], 
    "Education": ["Education_Attraction", "Education_Latitude", "Education_Longitude"],
    "Other": ["Other_Attraction", "Other_Latitude", "Other_Longitude"]
}

# Create the plots
fig, axes = plt.subplots(3, 2, figsize=(14, 12))  # Create 6 subplots
axes = axes.flatten()  # Flatten axes array to easily access each subplot

# Loop through each attraction type and plot the corresponding scatter plot
for idx, (attraction_type, columns) in enumerate(attraction_columns.items()):
    ax = axes[idx]
    
    # Plot metro stations as red markers
    for _, row in metro_dataset.dropna(subset=['Latitude', 'Longitude']).iterrows():
        ax.scatter(row['Longitude'], row['Latitude'], c='grey', label='Metro Station' if _ == 0 else "", alpha=0.6)
    
    # Plot attractions as color-coded markers for each type
    for _, row in metro_dataset.dropna(subset=[columns[1], columns[2]]).iterrows():
        ax.scatter(row[columns[2]], row[columns[1]], c=attraction_colors[attraction_type], 
                   label=attraction_type if _ == 0 else "", alpha=0.6)
    
    # Set title and labels
    ax.set_title(f"{attraction_type} Attractions and Metro Stations", fontsize=14)
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")
    
    # Add legend
    ax.legend(title="Legend", loc='upper left', fontsize=10)

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```

**Summary**

The graphs show how different attractions (Entertainment, Workplace, Health, Education, and Other) are located near metro stations. Each dot represents either a metro station or an attraction. You can see most attractions are close to metro stations, meaning theyâ€™re easy to reach. However, a few attractions are far away, making them harder to access.

You can see outliers, especially in Health and Other categories, where some attractions are much farther away. This means while most places are close, a few are significantly harder to reach.

```python
# Display the visualization
# Extract morning and night columns
morning_cols = [col for col in metro_dataset.columns if 'Morning' in col]
night_cols = [col for col in metro_dataset.columns if 'Night' in col]

# Calculate averages for plotting
morning_avg = metro_dataset[morning_cols].mean()
night_avg = metro_dataset[night_cols].mean()

# Line Chart
plt.figure(figsize=(10, 6))
plt.plot(morning_cols, morning_avg, label='Morning', marker='o')
plt.plot(night_cols, night_avg, label='Night', marker='o')
plt.title('Monthly Morning vs. Night Activity')
plt.xlabel('Month')
plt.ylabel('Average Activity')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

**Summary**

This graph compares morning and night activity levels across months. Morning activity peaks in June and drops sharply after September, while night activity starts lower, peaks in July, and declines towards December. The pattern shows more activity in the summer months for both times but with higher engagement in the morning. This suggests seasonal variation in behavior, with more outdoor activity during the summer.

```python
import pandas as pd
import matplotlib.pyplot as plt


# Define walkability columns
walkability_columns = [
    "Entertainment_Walkable(yes/no)", 
    "Workplace_Walkable(yes/no)", 
    "Health_Walkable(yes/no)", 
    "Education_Walkable(yes/no)", 
    "Other_Walkable(yes/no)"
]

# Handle missing values by filling 'NaN' with 'No' for walkability columns
metro_dataset[walkability_columns] = metro_dataset[walkability_columns].fillna('No')

# Calculate the value counts for each walkability column
walkability_df = metro_dataset[walkability_columns].apply(lambda x: x.value_counts())

# Print the value counts for better visualization
print("\nWalkability Analysis Results:")
print(walkability_df)

# Plotting the value counts for each walkability category
walkability_df.T.plot(kind='bar', figsize=(14, 7), color=['#1f77b4', '#ff7f0e'])

# Adding titles and labels
plt.title('Walkability Status for Metro Stations', fontsize=16)
plt.ylabel('Number of Stations', fontsize=12)
plt.xlabel('Walkability Categories', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Walkable', loc='upper left')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

**Summary**

This graph shows the walkability status of metro stations for different categories of attractions like Entertainment, Workplace, Health, Education, and Other. The blue bars represent the stations that are not walkable, while the orange bars show the walkable ones. Most stations are not walkable, especially for Education and Health attractions. However, the Other category has nearly equal numbers of walkable and non-walkable stations.
